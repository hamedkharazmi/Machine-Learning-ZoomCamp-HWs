```python
import pandas as pd
import requests
import json
```

```python
df = pd.read_csv('bank-additional-full.csv', sep=';')
df.columns = df.columns.str.lower().str.replace(" ", "_")

binary_values = {
    'no': 0,
    'yes': 1
}
df["y"] = df["y"].map(binary_values)

del df['loan']
del df['housing']
del df['day_of_week']
del df['marital']
del df['education']
del df['default']
```

```python
samples = df.sample(n=5)
samples
```

```python
# url = 'http://localhost:9696/predict'
url = 'https://bank-marketing-system.herokuapp.com/predict'

for sample in samples.apply(lambda x: json.loads(x.to_json()), axis=1):
    val = sample.pop('bank-marketing-system', None)
    response = requests.post(url, json=sample).json()
    print(response, val)
```
